```python
#Data visualization for all the hotspots of Toronto. File Downloaded from: 
#https://open.toronto.ca/dataset/places-of-interest-and-toronto-attractions/
#using Jupitor Notebook

import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import folium
from shapely.geometry import Point
import json

#load the hotspots file:
data = pd.read_csv('Places of Interest and Attractions - 4326.csv')

#get all the names of places
place_names = data.NAME

#file specific edit
geometry = data['geometry'].str.replace("'", '"')

#get Coordinates
coordinates= geometry.apply(lambda x: json.loads(x)['coordinates'][0])

# Load the natural earth dataset for cities
cities = gpd.read_file(gpd.datasets.get_path("naturalearth_cities"))

# Filter the dataset to get only Toronto
toronto = cities[cities['name'] == 'Toronto']
toronto_location = [toronto.geometry.y.values[0], toronto.geometry.x.values[0]]

#start with Toronto Location
my_map = folium.Map(location=toronto_location, zoom_start=12)

#Creating a list with all the points
for coordinate, name in zip(coordinates, place_names):
        #Not a mistake file adjustment
        longitude, latitude = coordinate
        folium.Marker(location=[latitude, longitude], popup="popup", tooltip=name).add_to(my_map)


#using Jupitor Notebook
my_map
```
